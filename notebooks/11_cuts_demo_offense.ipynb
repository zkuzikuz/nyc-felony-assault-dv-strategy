# 11 — Demographic, Offense, Operational Cuts (C13-C15, C19-C20)

Categorical cuts (complaint types, law category, top felonies) plus operational
cuts (repeat-address 311, agency response time). C7-C12 (NTA / tract / ACS) and
C18 (GDELT) are run inline in the H1/H4 hypothesis notebooks.


```python
import sys
from pathlib import Path
PROJECT = Path.cwd().parent if Path.cwd().name == 'notebooks' else Path.cwd()
sys.path.insert(0, str(PROJECT / 'src'))

import matplotlib.pyplot as plt
from nyc_eda import DUCKDB_PATH, CUTS_OUT
from nyc_eda.storage import open_db
from nyc_eda import cuts

con = open_db(DUCKDB_PATH)
```

## C13 — Top 311 complaint types


```python
df = cuts.c13_311_complaint_types(con)
df.to_csv(CUTS_OUT / 'C13_311_top.csv', index=False); df.head(25)
```

## C14 — Law category split


```python
df = cuts.c14_law_category(con)
df.to_csv(CUTS_OUT / 'C14_law_category.csv', index=False); df
```

## C15 — Top felony offenses, year over year


```python
df = cuts.c15_top_felony_offenses(con)
df.to_csv(CUTS_OUT / 'C15_top_felony.csv', index=False)
wide = df.pivot(index='year', columns='ofns_desc', values='n').fillna(0)
fig, ax = plt.subplots(figsize=(10, 5))
wide.plot(ax=ax, marker='o'); ax.set_title('C15 — top felony offenses by year')
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5), fontsize=8)
plt.tight_layout(); plt.savefig(CUTS_OUT / 'C15_top_felony.png', dpi=120); plt.show()
```

## C19 — Repeat-address 311 hotspots


```python
df = cuts.c19_repeat_address_311(con)
df.to_csv(CUTS_OUT / 'C19_repeat_addr.csv', index=False); df.head(30)
```

## C20 — Resolution time by agency


```python
df = cuts.c20_resolution_time_by_agency(con)
df.to_csv(CUTS_OUT / 'C20_resolution_time.csv', index=False); df
```
