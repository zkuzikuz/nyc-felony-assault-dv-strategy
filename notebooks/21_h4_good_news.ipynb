## Felony trends 2019 → latest year


```python
df = cuts.c15_top_felony_offenses(con, top_n=15)
wide = df.pivot(index='year', columns='ofns_desc', values='n').fillna(0)
if 2019 in wide.index and wide.index.max() >= 2024:
    base = wide.loc[2019]; latest = wide.loc[wide.index.max()]
    pct = ((latest - base) / base * 100).sort_values()
    print('Pct change 2019 → latest, by offense:'); print(pct.round(1))
else:
    print('Need both 2019 and >=2024 in data; current range:', wide.index.min(), '–', wide.index.max())
```
