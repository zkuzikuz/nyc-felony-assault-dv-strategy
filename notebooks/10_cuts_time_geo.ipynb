# 10 — Time and Geography Cuts (C1-C6)

Runs cuts C1 (calendar year), C2 (seasonality), C3 (COVID step), C4 (DOW × hour),
C5 (borough), C6 (precinct). Saves figures to `data/processed/cuts/`.


```python
import sys
from pathlib import Path
PROJECT = Path.cwd().parent if Path.cwd().name == 'notebooks' else Path.cwd()
sys.path.insert(0, str(PROJECT / 'src'))

import matplotlib.pyplot as plt
from nyc_eda import DUCKDB_PATH, CUTS_OUT
from nyc_eda.storage import open_db
from nyc_eda import cuts

CUTS_OUT.mkdir(parents=True, exist_ok=True)
con = open_db(DUCKDB_PATH)
```

## C1 — Calendar year


```python
df = cuts.c1_calendar_year(con)
df.to_csv(CUTS_OUT / 'C1_calendar_year.csv', index=False)
fig, ax = plt.subplots(figsize=(8, 4))
for src, sub in df.groupby('source'):
    ax.plot(sub['year'], sub['n'], marker='o', label=src)
ax.set_title('C1 — complaints per calendar year'); ax.set_xlabel('year'); ax.legend()
plt.tight_layout(); plt.savefig(CUTS_OUT / 'C1_calendar_year.png', dpi=120); plt.show()
df
```

## C2 — Seasonality


```python
df = cuts.c2_month_seasonality(con)
df.to_csv(CUTS_OUT / 'C2_seasonality.csv', index=False)
fig, ax = plt.subplots(figsize=(8, 4))
ax.bar(df['month'], df['avg_complaints'])
ax.set_title('C2 — NYPD avg monthly complaints (across years)')
ax.set_xlabel('month'); ax.set_ylabel('avg complaints')
plt.tight_layout(); plt.savefig(CUTS_OUT / 'C2_seasonality.png', dpi=120); plt.show()
df
```

## C3 — COVID step


```python
df = cuts.c3_covid_step(con)
df.to_csv(CUTS_OUT / 'C3_covid_step.csv', index=False)
df
```

## C4 — Day-of-week × hour


```python
df = cuts.c4_dow_hour(con)
df.to_csv(CUTS_OUT / 'C4_dow_hour.csv', index=False)
pivot = df.pivot(index='dow', columns='hour', values='n').fillna(0)
fig, ax = plt.subplots(figsize=(10, 3.5))
im = ax.imshow(pivot.values, aspect='auto', cmap='YlOrRd')
ax.set_yticks(range(7)); ax.set_yticklabels(['Sun','Mon','Tue','Wed','Thu','Fri','Sat'])
ax.set_xlabel('hour of day'); ax.set_title('C4 — NYPD complaints: DOW × hour')
fig.colorbar(im); plt.tight_layout(); plt.savefig(CUTS_OUT / 'C4_dow_hour.png', dpi=120); plt.show()
```

## C5 — Borough


```python
df = cuts.c5_borough(con)
df.to_csv(CUTS_OUT / 'C5_borough.csv', index=False)
df
```

## C6 — Precinct


```python
df = cuts.c6_precinct(con)
df.to_csv(CUTS_OUT / 'C6_precinct.csv', index=False)
df.head(20)
```
